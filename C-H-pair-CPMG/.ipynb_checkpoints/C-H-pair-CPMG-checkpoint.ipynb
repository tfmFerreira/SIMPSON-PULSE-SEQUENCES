```python
#Testing CPMG on a powder of 13C-1H spin pairs (SIMPSON)  
#Tiago M. Ferreira

from matplotlib import pyplot as plt
import numpy as np
```

```python
c=np.loadtxt('C-H-pair-Pake_TD_dw=10microsec_NP=512.dat')
x3=c[:,0]
y3=c[:,1]*np.exp(-lb*x3)
fig=plt.figure()
plt.plot(x3,y3,'k-',linewidth=1,label='Simple FID (no pulses)')

plt.xlabel("time / seconds")
plt.ylabel("$^{13}$C magnetisation")

print(f"FID with no pulses")
```

```python
lb=200

a=np.loadtxt('C-H-pair-CPMG_TD_dw=10microsec_btwpulses=320microsec_NP=512.dat')
x1=a[:,0]
y1=a[:,1]*np.exp(-lb*x1)
fig=plt.figure()
plt.plot(x1,y1,'b-',linewidth=1,label='320 microsec')
#plt.plot(x1,y1,'bo',linewidth=1)
#ax = plt.gca()
#ax.set_xlim([0, 0.001])

plt.xlabel("time / seconds")
plt.ylabel("$^{13}$C magnetisation")

plt.legend()

b=np.loadtxt('C-H-pair-CPMG_TD_dw=10microsec_btwpulses=640microsec_NP=512.dat')
x2=b[:,0]
y2=b[:,1]*np.exp(-lb*x2)
fig=plt.figure()

plt.plot(x2,y2,'r-',linewidth=1,label='640 microsec')

plt.xlabel("time / seconds")
plt.ylabel("$^{13}$C magnetisation")
plt.legend()

print(f"Time domain using 180 pulses spaced by the delays indicated (CPMG).")


```

```python
fig=plt.figure()




y1[0]=0.5*y1[0]
F1 = np.fft.fft(y1)
freq1 = np.fft.fftfreq(x1.size, d=x1[1]-x1[0])
F_shifted1 = np.fft.fftshift(F1)
freq_shifted1 = np.fft.fftshift(freq1)

plt.plot(freq_shifted1,np.real(F_shifted1)+50,'b-',label='320 microsec')

y2[0]=0.5*y2[0]
F2 = np.fft.fft(y2)
freq2 = np.fft.fftfreq(x2.size, d=x2[1]-x2[0])
F_shifted2 = np.fft.fftshift(F2)
freq_shifted2 = np.fft.fftshift(freq2)

plt.plot(freq_shifted2,np.real(F_shifted2)+10,'r-',label='640 microsec')

y3[0]=0.5*y3[0]
F3 = np.fft.fft(y3)
freq3 = np.fft.fftfreq(x3.size, d=x3[1]-x3[0])
F_shifted3 = np.fft.fftshift(F3)
freq_shifted3 = np.fft.fftshift(freq3)

plt.plot(freq_shifted3,np.real(F_shifted3),'k-',label='No pulses (Pake)')

plt.xlabel("frequency / Hz")
plt.ylabel("Intensity")
plt.legend()

print(f"Frequency domain")

frame1 = plt.gca()
frame1.axes.get_yaxis().set_visible(False)
```
