```python
#Testing CPMG on a powder of 13C-1H spin pairs (SIMPSON)  
#Tiago M. Ferreira

from matplotlib import pyplot as plt
import numpy as np
```

```python
lb=200

a=np.loadtxt('C-H-pair-CPMG_TD_dw=4microsec_btwpulses=64microsec_NP=512.dat')
x1=a[:,0]
y1=a[:,1]*np.exp(-lb*x1)
fig=plt.figure()
plt.plot(x1,y1,'b-',linewidth=1)
#plt.plot(x1,y1,'bo',linewidth=1)
ax = plt.gca()
#ax.set_xlim([0, 0.001])

b=np.loadtxt('C-H-pair-CPMG_TD_dw=4microsec_btwpulses=128microsec_NP=512.dat')
x2=b[:,0]
y2=b[:,1]*np.exp(-lb*x2)
fig=plt.figure()

plt.plot(x2,y2,'r-',linewidth=1)
```

```python
y1[0]=0.5*y1[0]
F1 = np.fft.fft(y1)
freq1 = np.fft.fftfreq(x1.size, d=x1[1]-x1[0])
F_shifted1 = np.fft.fftshift(F1)
freq_shifted1 = np.fft.fftshift(freq1)

fig=plt.figure()
plt.plot(freq_shifted1,np.real(F_shifted1),'b-')

y2[0]=0.5*y2[0]
F2 = np.fft.fft(y2)
freq2 = np.fft.fftfreq(x2.size, d=x2[1]-x2[0])
F_shifted2 = np.fft.fftshift(F2)
freq_shifted2 = np.fft.fftshift(freq2)

plt.plot(freq_shifted2,-np.real(F_shifted2),'r-')
```
