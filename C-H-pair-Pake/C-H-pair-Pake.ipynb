```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
a=np.loadtxt('tmp_TD.dat')
x=a[:,0]
y=a[:,1]
fig=plt.figure()
#plt.plot(traj.time[:],corr[:])
plt.plot(x,y)
plt.plot(x,y,'bo')
```

```python
y[0]=0.5*y[0]
F = np.fft.fft(y)
freq = np.fft.fftfreq(x.size, d=x[1]-x[0])

F_shifted = np.fft.fftshift(F)
freq_shifted = np.fft.fftshift(freq)

fig=plt.figure()
#plt.plot(traj.time[:],corr[:])
plt.plot(freq_shifted,np.real(F_shifted))
```
